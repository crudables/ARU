### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

%matplotlib inline
```

### Step 2. Import the dataset and Assign it to a variable titanic 

```python

titanic = pd.read_csv('titanic.csv')

titanic.head()
```

```python
titanic.info()
```

```python
titanic.shape
```

```python
titanic.columns = titanic.columns.str.lower()
```

### Step 3. Set PassengerId as the index 

```python
titanic.set_index('passengerid',inplace=True)
```

### Step 4. Create a pie chart presenting the male/female proportion

```python
r =  titanic.groupby('sex')['sex'].count()
```

```python
plt.pie(x=r,labels=['female','male'],autopct='%.0f%%')
plt.show()
```

### Step 5. Create a scatterplot with the Fare payed and the Age, differ the plot color by gender

```python
sns.scatterplot(y=titanic['fare'], x=titanic['age'],hue=titanic['sex'])
plt.show()
```

### Step 6. How many people survived?

```python
titanic.query("survived == 1")['survived'].count()
```

### Step 7. Create a histogram with the Fare payed

```python
sns.histplot(x=titanic['fare'],bins=15)
plt.yscale('log')
# plt.show()
```

### BONUS: Create your own question and answer it.

## According to the scatterplot above, it seems female paid higher fare. Is this true?

```python
t_fare = titanic.groupby(by='sex').fare.mean()
t_fare = pd.DataFrame(t_fare)
```

```python
t_fare.reset_index(level=0,inplace=True)
```

```python
# plt.bar(data=t_fare,x = 'sex',height='fare',width=0.5)
sns.barplot(data=t_fare, x='sex',y='fare').set(title='Average fare prices for male and female')
```

**It is now obvious that females on average paid higher fees that males**
